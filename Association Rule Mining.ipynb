# Association Rule Mining

## Importing Libraries

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.preprocessing import OrdinalEncoder
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

## Reading Data for arules 1 and Data for arules 2

```python
arules1=pd.read_csv("C:/Data/Data for arules 1.csv")
```

```python
arules2=pd.read_csv("C:/Data/Data for arules 2.csv")
```

## 1. Association Rule for Data for arules 1 :

```python
arules1.head()
```

### Removing first 4 columns

```python
newarules1 = pd.DataFrame(arules1)
newarules1.drop(["resid","familysz","chb18","cha18"],axis=1,inplace=True)
newarules1.head()
```

### Ordinal Encoder () used to convert "Yes" and "No" to 1 and 0 and formed an array

```python
# Converting the yes and no values to 1 and 0 respectively.
transformation = OrdinalEncoder()
newarules_1 = transformation.fit_transform(newarules1)
newarules_1
```

```python
df=pd.DataFrame(newarules_1)
df
```

### Creating dataframe of dataset from array above with column names

```python
new_df1=df.rename(columns={0: "borvita", 1: "milo",2:'zandu',3:'dred',4:'dblue',5:'prtnex',6:'holik',7:'badya',8:'complan'})
new_df1
```

## Creating table for frequent items bought with minimum Support=30%
### Support refers to the popularity of a product out of total transactions. Here we have considered min support to be 30% and more

```python
freq_items1=apriori(new_df1,min_support=0.3,use_colnames=True)
freq_items1
```

## Next Rule applied is for Confidence where minimum Confidence is 30%
### Confidence is the probability of buying product 2 if product 1 is bought. Below we have considered the rule where confidence is 30% and more. For below dataset most of the items ahve confidence level 100%.

```python
rules=association_rules(freq_items1,metric="confidence",min_threshold=0.3)
rules
```

## Next Rule is according to Lift threshold where minimum Lift applied is for 2.1.
### Lift is a measure of correlation. Lift needs to be greater than 1 for the probability of 2 products to be bought together. For above dataset minimum Lift value is 1.3 and we are filetring for graeter than 2.1.

```python
rules1=association_rules(freq_items1,metric="lift",min_threshold=2.1)
rules1
```

## Calculating length of Antecedents columns to check no. of products

```python
rules1["antecedents_len"]=rules1["antecedents"].apply(lambda x:len(x))
rules1
```

## Final rule applied where Confidence is 100% and Lift is greater than 2.3
### Outcome:
This show that if a customer buys prtnex and holik, then there is 100% probability that they will buy compaln also as the Confidence is 1.0.

```python
rules1[(rules1["antecedents_len"]>=2) & (rules1["confidence"]>=1) & (rules1["lift"]>2.3)]
```

## 2. Association Rule for Data for arules 2 :

```python
arules2.head()
```

### Removing irrelevant column Row Labels 

```python
newarules2 = pd.DataFrame(arules2)
newarules2.drop(["Row Labels"],axis=1,inplace=True)
newarules2.head()
```

### Checking for Null values if any

```python
newarules2.info()
```

### Replaced Null values with 0 as Apriori alogorithm works on either 1 and 0 or True and False

```python
newarules2.fillna(0,inplace=True)
newarules2.info()
```

```python
newarules2.head()
```

### Replacing values with data 2,3 and 4 with 1 

```python
newarules2.replace([2,3,4],1,inplace=True)
```

```python
newarules2.head()
```

## Creating table for frequent products with minimum Support=30%

```python
freq_items2=apriori(newarules2,min_support=0.3,use_colnames=True)
freq_items2
```

## Next Rule is applied for Confidence with minimum threshold of 50%

```python
rules=association_rules(freq_items2,metric="confidence",min_threshold=0.5)
rules
```

## Next Rule applied for Lift with minimum threshold of 1.4.
### From below result it is clear which products are most popular with Lift>1

```python
rules1=association_rules(freq_items2,metric="lift",min_threshold=1.6)
rules1
```

## Calculating the length of Antecedent column to check no. of products

```python
rules1["antecedents_len"]=rules1["antecedents"].apply(lambda x:len(x))
rules1
```

## Final rule is applied for Confidence level 78% and Lift > 1.7
### Outcome:
Below result shows that if a customer buys Banana and Beans then there is 78.9% probability that they will buy Root Vegetables also as the Confidence is 78.9%. According to the analysis the products can be arranged in different patterns in store for maximum sale.

```python
rules1[(rules1["antecedents_len"]>=2) & (rules1["confidence"]>=0.7) & (rules1["lift"]>1.7)]
```
